# Count the movies in each genre

```python
from collections import Counter
cnt = Counter(df.Action)
cnt1 = Counter(df.Animation)
cnt2 = Counter(df.Comedy)
cnt3 = Counter(df.Drama)
cnt4 = Counter(df.Documentary)
cnt5 = Counter(df.Romance)
cnt6 = Counter(df.Short)
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
import pandas as pd

raw_data = { 'Action ':[cnt.get(1)], 'Animation':[cnt1.get(1)], 'Comedy':[cnt2.get(1)], 'Drama':[cnt3.get(1)], 'Documentry':[cnt4.get(1)], 'Romance':[cnt5.get(1)],'short':[cnt6.get(1)]}
```

```python
df1= pd.DataFrame(raw_data)
```

```python
df1 = df1.T
```

```python
df1.columns=[1]
df1
```

```python
fim = df1.plot(kind="bar", color='orange')
fim.legend().set_visible(False)
plt.ylabel('Frequency')
plt.xlabel('Genre')
plt.title('Genre of the Movies', color= 'brown')
```

# Regression Analysis

```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy as stats
import matplotlib.pyplot as plt
import sklearn

df_movies = pd.DataFrame(dfe, columns=dfe.columns)
```

```python
df_movies
```

```python
from sklearn.linear_model import LinearRegression

y = df_movies['rating']
x = df_movies.drop(['rating','Unnamed: 0','title','mpaa','r1','r2','r3','r4','r5','r6','r7','r8','r9','r10'], axis =1).as_matrix()
```

```python
x
```

```python
import seaborn as sns
sns.pairplot(df_movies, vars=['rating','year','length','budget','votes'])
```

```python
import seaborn as sns
sns.pairplot(df_movies, vars=['rating','Action','Animation','Comedy','Drama','Romance','Documentary','Short'])
```

```python
import seaborn as sns
sns.pairplot(df_movies, vars=['rating','budget','votes'])
```

```python
lm = LinearRegression()
lm.fit(x,y)
```

```python
lm.predict(x)
```

```python
names = df_movies.drop(['rating','Unnamed: 0','title','mpaa','r1','r2','r3','r4','r5','r6','r7','r8','r9','r10'], axis =1)
coefs = list(zip(lm.coef_,names.columns))
```

```python
lm.intercept_
```

```python
coefs
```

```python
print('Variance score: {}'.format(lm.score(x, y)))
```

```python
plt.scatter(y,lm.predict(x))
```

```python
import statsmodels.api as sm 
from statsmodels.sandbox.regression.predstd import wls_prediction_std

model1 = sm.OLS(y,sm.add_constant(x))
```

```python
result = model1.fit()
```

```python
result.summary()
```

```python
error = sklearn.metrics.mean_squared_error(lm.predict(x),y)
```

```python
print('error is {}'.format(error))
```

```python
plt.scatter(lm.predict(x),lm.predict(x)-y,alpha=0.5, color='blue')
plt.hlines(y = 0, xmin = 0, xmax = 25, linewidth = 2)
plt.title('Residual Plot', color='red')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
```
